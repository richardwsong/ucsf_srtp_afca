```python
# read in the list of genes 

import sys
import os
import anndata
import numpy as np

gene_list = []
with open("nutrient_genes.txt", 'r') as f:
    for line in f:
        gene_list.append(line.strip())

adata = anndata.read_h5ad('adata_head_S_v1.0.h5ad') 

# get the list of genes that are in the adata object
adata_genes = adata.var_names
adata_filter = [gene for gene in gene_list if gene in adata_genes]
cell_filter = adata.obs['afca_annotation'] == 'ensheathing glial cell'
# filter the adata object
adata_filtered = adata[cell_filter, adata_filter]
```

```python
import plotly.express as px
import pandas as pd

X = np.array(adata_filtered.X.toarray())
ages = np.array(adata_filtered.obs['age'].values, dtype=int)

# pca on X 
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(X)
X_pca = pca.transform(X)

# Create a DataFrame from the PCA results
df = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3'])
df['Age'] = ages

# Create a 3D scatter plot
fig = px.scatter_3d(df, x='PC1', y='PC2', z='PC3', color='Age', size_max=.1)
fig.show()
```
